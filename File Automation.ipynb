```python
import fnmatch
import re
import os
import shutil
from docx import Document
import logging as lg
import sys

os.chdir("D:\\Ineuron Assignment")

formatter = lg.Formatter("%(asctime)s - %(name)s - %(levelname)s - %(message)s")
LOG_FILe = "File Automation.log"


def logging(text):
    """This function will log the operation at different stage and create a file accordingly
    to display various operation with respect to timestamp"""
    user = lg.getLogger("User")
    file_handler = lg.FileHandler(LOG_FILe)
    stream_handler = lg.StreamHandler(sys.stdout)
    file_handler.setFormatter(formatter)
    stream_handler.setFormatter(formatter)
    user.setLevel(lg.INFO)
    user.addHandler(file_handler)
    user.addHandler(stream_handler)
    user.info(text)


def fileselection(path):
    """This function helps to select only required files and keeping unnecessary files untouched"""

    file_list = []
    for files in path:
        if os.path.isfile(files):
            if fnmatch.fnmatch(files, "*ment*"):
                file_list.append(files)
    return file_list
    logging("File selection is done")


def folder_creation():
    """This function helps to create folder with respect to common file name"""

    # Finding common name and manipulation
    temp1 = []
    temp2 = []

    # Removing extension from files
    for file in os.listdir():
        if os.path.isfile(file):
            if file in fileselection(os.listdir()):
                temp1.append(os.path.splitext(file)[0])

    # removing all characters except letters
    for files in temp1:
        temp2.append("".join(re.findall("[a-zA-Z]", files)))

    # Taking common file name in set to create folder as per unique file names
    folder_name = list(set(temp2))

    # Creating folder/folders based on number of common files in the list
    for folder in folder_name:
        if not os.path.isdir(folder):
            os.mkdir(os.path.join(os.getcwd(), folder))
            logging("Folder selection is done")
            logging("Folder name is" + str(folder))

    return folder_name


def rename_file(path):
    """This function helps to rename files by excluding all characters except letters and numbers
     & returns original file name by excluding extension part """
    file_name = []
    for i in path:
        if os.path.isfile(i):
            if i in fileselection(os.listdir()):
                file_name.append((i.split(".")[0]))
                os.rename(i, (re.sub('[\W_]+', '', (i.split(".")[0]))) + ".docx")
    logging("File renaming is done")
    return file_name


def move_file(path):
    """This function helps to move desired files to respective folders"""

    for i in path:
        if os.path.isdir(i):
            if i in folder_creation():
                for j in rename_file(path):
                    if i == re.sub(r'[^A-Za-z]', "", j):
                        shutil.move(os.getcwd() + "\\" + j + ".docx", os.getcwd() + "\\" + i)
    logging("All files are moved to respective directory")


def merge_file(paths):
    """This function helps to merge all docx file into a single docx file and separating all files
    by means of file name as header in the merged document"""

    master_file = Document()
    for i in paths:
        if i in fileselection(paths):
            if os.path.isfile(i):
                source = Document(i)
                master_file.add_heading(re.sub('[\W_]+', '', i.split(".")[0]))
                for paragraph in source.paragraphs:
                    text = paragraph.text
                    master_file.add_paragraph(text)
                shutil.move(os.getcwd() + "\\" + i, os.getcwd() + "\\" + re.sub(r'[^A-Za-z]', '', i.split(".")[0]))
    master_file.save("master file.docx")
    logging("File merging operation is complete")


# step 1- Create folders based on common file name and selecting only required files
folder_creation()
# step 2- rename all file names
rename_file(os.listdir())
# step-3 Merge all the files to a single file
merge_file(os.listdir())
# step-4 Move all the files to allocated location based on file name
move_file(os.listdir())
```
